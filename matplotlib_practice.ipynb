```python
%matplotlib inline
```

```python
# Import dependencies

import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.

x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to the list of floats as the total fare in US dollars accumulated each month.

y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.05]
```

```python
# Create the plot

plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()

fig, ax = plt.subplots()

ax.plot(x_axis, y_axis)
```

```python
fig = plt.figure()

ax = fig.add_subplot()

ax.plot(x_axis, y_axis)
```

```python
# Create a plot with the ax.plt()

ax = plt.axes()

ax.plot(x_axis, y_axis)
```

```python
# Create the plot.

plt.plot(x_axis, y_axis)

plt.show()
```

```python
# Annotating charts

# Create the plot and add a label for the legend.

plt.plot(x_axis, y_axis, label ='Boston')

# Create labels for the x and y axes.

plt.xlabel("Date")

plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.

plt.ylim(0, 45)

# Create the title.

plt.title("PyBer Fare by Month")

# Add the legend.

plt.legend()
```

```python
# Amending code block to declare parameters to change how the plt.plot() function will look.

plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')

# Create labels for the x and y axes.

plt.xlabel("Date")

plt.ylabel("Fares($)")

# Set the y limit between 0 and 45.bit_length
plt.ylim(0, 45)

# Create the title.

plt.title("PyBer Fare by Month")

# Add a grid.

plt.grid()

# Add the legend.

plt.legend()
```

```python
# Amending code block to declare parameters to change how the plt.plot() function will look.

fig, ax = plt.subplots()

ax.plot(x_axis, y_axis, marker="d", color="green", linewidth=2, label='Boston')

# Create labels for the x and y axes.

ax.set_xlabel('Date')

ax.set_ylabel('Fares($)')

# Set the y limit between 0 and 45.bit_length
ax.set_ylim(0, 45)

# Create the title.

ax.set_title("PyBer Fare by Month")

# Add a grid.

ax.grid()

# Add the legend.

ax.legend()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot for a bar chart

plt.bar(x_axis, y_axis)
```

```python
# Annotate the plot

plt.bar(x_axis, y_axis, color="green", label='Boston')

# Create labels for the x and y axes.

plt.xlabel("Date")

plt.ylabel("Fare($)")

# Create a title.

plt.title("PyBer Fare by Month")

# Add the legend.

plt.legend()
```

```python
# Create the plot for a horizontal bar chart

plt.barh(x_axis, y_axis)
```

```python
# Opposite axes

plt.barh(y_axis, x_axis)
```

```python
# Invert y-axis to have the months in ascending order.

plt.barh(x_axis, y_axis, color='magenta', label="Boston")

plt.gca().invert_yaxis()

# Skill Drill

plt.xlabel("Fare($)")

plt.ylabel("Date")

plt.title("PyBer Fare by Month")

plt.legend()
```
